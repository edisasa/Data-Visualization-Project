```python
import pandas as pd
import numpy as np
import matplotlib as pl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
MyData = pd.read_csv('Denver_Crime_2001-2013.csv', sep=",")
MyData
```

```python
WorkingData = MyData.copy() # Just to make sure we have some backup. 
WorkingData.set_index('Type', inplace=True)
WorkingData
```

```python
TransData = WorkingData.transpose()
TransData
```

```python
sns.set() 
%matplotlib inline
#%load_ext birdseye


x = list(TransData.index) # With long holiday names, this column displays better on the y-axis.
y = list(TransData.Murders) # This axis displays the traffic volume. 


plt.scatter(x,y)
plt.xlabel("Year")
plt.ylabel("Murder Cases")
plt.title("Year 2013")
```
