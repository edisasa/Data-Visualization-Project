```python
# Importing important libraries including MatPlotLib and its subset library.
import pandas as pd
import numpy as np
import matplotlib as pl
import matplotlib.pyplot as plt
```

#### Importing the first dataset

```python
FirstData = pd.read_csv('Denver_Crime_2001-2013.csv', sep=",")
WorkingData = FirstData.copy() # Just to make sure we have some backup. 
WorkingData.set_index('Type', inplace=True) #Made the first column (Type) an index column.
WorkingData.head() # Displays the first five observations.
```

#### Transposing table for analysis

```python
TransData = WorkingData.transpose() #Transposed the table to better analyze the data.
TransData.head() # Displays the first five observations.
```

#### Creating a Line Plot with subplots for the eight different crimes observed between 2001 through 2013

```python
x = list(TransData.index) #This line will use the Index Colument (Type) which is years 2001-2013.
y = list(TransData.Murders) # This axis displays the number of murders. 
z = list(TransData.Rapes) # This axis displays the number of rapes. 
a = list(TransData.Robberies) # This axis displays the number of robberies. 
b = list(TransData.Assaults) # This axis displays the number of assults. 
c = list(TransData.Burglaries) # This axis displays the number of Burglaries. 
d = list(TransData.Thefts) # This axis displays the number of thefts. 
e = list(TransData.AutoThefts) # This axis displays the number of Auto Thefts. 
f = list(TransData.Arson) # This axis displays the number of arsons. 


fig = plt.figure(figsize=(15,8)) # Sets the Figure size.
fig.tight_layout() # Provides space between subplots so that lables will not overlap.
fig.set_facecolor('None')# removes the background color of the Figure.
fig.suptitle('Figure 1: Denver Crime Patterns from 2001-2013 ', fontsize=20) # Sets Figure tile.

axis1 = fig.add_subplot(241) # Creates the first subplot in a 2-rows, 4-column Figure.
axis1.plot(x,y, label="Murders", linewidth=1.5) # Creates a plot with lable and sets the line width.
axis1.set_title('Murders', fontsize=14) # Sets the title of the subplot.
axis1.set_xlabel('Year', fontsize=12) # Sets the lable for the x-axis of the subplot.
axis1.tick_params(axis='x', labelrotation=90) # Rotates the x-axis ticks into vertical position.
axis1.set_ylabel('Number of Cases', fontsize=12, labelpad=3) # Sets the y-axis label.
axis1.set_facecolor('None') # Removes the background color of the subplot.
axis1.xaxis.grid(False) # Removes gridlines along the x-axis.
axis1.yaxis.grid(False) # Removes gridlines along the y-axis.

#The above comments similarly describe the corresponding lines of code the remaining 7 subplots.

axis2 = fig.add_subplot(242)
axis2.plot(x,z, label="Rapes", linewidth=1.5)
axis2.set_title('Rapes', fontsize=14)
axis2.set_xlabel('Year', fontsize=12)
axis2.tick_params(axis='x', labelrotation=90)
axis2.set_ylabel('Number of Cases', fontsize=12, labelpad=3)
axis2.set_facecolor('None')
axis2.xaxis.grid(False)
axis2.yaxis.grid(False)

axis3 = fig.add_subplot(243)
axis3.plot(x,a, label="Robberies", linewidth=1.5)
axis3.set_title('Robberies', fontsize=14)
axis3.set_xlabel('Year', fontsize=12)
axis3.tick_params(axis='x', labelrotation=90) 
axis3.set_ylabel('Number of Cases', fontsize=12, labelpad=3)
axis3.set_facecolor('None')
axis3.xaxis.grid(False)
axis3.yaxis.grid(False)

axis4 = fig.add_subplot(244)
axis4.plot(x,b, label="Assualts", linewidth=1.5)
axis4.set_title('Assualts', fontsize=14)
axis4.set_xlabel('Year', fontsize=12)
axis4.tick_params(axis='x', labelrotation=90)
axis4.set_ylabel('Number of Cases', fontsize=12, labelpad=3)
axis4.set_facecolor('None')
axis4.xaxis.grid(False)
axis4.yaxis.grid(False)

axis5 = fig.add_subplot(245)
axis5.plot(x,c, label="Burglaries", linewidth=1.5)
axis5.set_title('Burglaries', fontsize=14)
axis5.set_xlabel('Year', fontsize=12)
axis5.tick_params(axis='x', labelrotation=90)
axis5.set_ylabel('Number of Cases', fontsize=12, labelpad=3)
axis5.set_facecolor('None')
axis5.xaxis.grid(False)
axis5.yaxis.grid(False)

axis6 = fig.add_subplot(246)
axis6.plot(x,d, label="Thefts", linewidth=1.5)
axis6.set_title('Thefts', fontsize=14)
axis6.set_xlabel('Year', fontsize=12)
axis6.tick_params(axis='x', labelrotation=90)
axis6.set_ylabel('Number of Cases', fontsize=12, labelpad=3)
axis6.set_facecolor('None')
axis6.xaxis.grid(False)
axis6.yaxis.grid(False)

axis7 = fig.add_subplot(247)
axis7.plot(x,e, label="Auto Thefts", linewidth=1.5)
axis7.set_title('Auto Thefts', fontsize=14)
axis7.set_xlabel('Year', fontsize=12)
axis7.tick_params(axis='x', labelrotation=90)
axis7.set_ylabel('Number of Cases', fontsize=12, labelpad=3)
axis7.set_facecolor('None')
axis7.xaxis.grid(False)
axis7.yaxis.grid(False)

axis8 = fig.add_subplot(248)
axis8.plot(x,f, label="Arson", linewidth=1.5)
axis8.set_title('Arson', fontsize=14)
axis8.set_xlabel('Year', fontsize=12)
axis8.tick_params(axis='x', labelrotation=90) 
axis8.set_ylabel('Number of Cases', fontsize=12, labelpad=3)
axis8.set_facecolor('None')
axis8.xaxis.grid(False)
axis8.yaxis.grid(False)
fig.tight_layout()
fig.savefig('FirstPlot.jpg', dpi=300)
```

#### Second Data Source

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00492/Metro_Interstate_Traffic_Volume.csv.gz"
SecondData = pd.read_csv(url, sep=",") # Data found online UCI server.
# I can also use this line below if there is no internet connection and data is available.
# SecondData = pd.read_csv(url, sep=",")
SecondData.head()
```

```python
Holiday = SecondData.groupby(['holiday'])[['traffic_volume']]. \
agg(np.mean)
Holiday 
```

#### Creating a Bar Plot 

```python
x = list(Holiday.index) # With long holiday names, this column displays better on the y-axis.
y = list(Holiday.traffic_volume) # This axis displays the traffic volume. 

fig, ax = plt.subplots(figsize=(15,8))
fig.tight_layout() # Provides space between subplots so that lables will not overlap.
fig.subplots_adjust(top=0.95, left=0.2) # I have been running into images being cropped out when saved. 
# This above line provides padding to the figure so that all labels and axes are visible. 

ax.bar(x, y)

ax.set_title('Traffic Volume on Holidays', fontsize=18)
ax.set_xlabel('Traffic Volume', fontsize=12)
ax.set_ylabel('Holiday', fontsize=12, labelpad=1.5)
ax.tick_params(axis='x', labelrotation=90) 
ax.set_facecolor('None') # Removes the background color of the subplot.
ax.xaxis.grid(False)
ax.yaxis.grid(False)

fig.savefig('SecondPlot.jpg', dpi=300)

```

#### Grouping Data by weather type to learn if traffic pattern changes occur.

```python
Weather = SecondData.groupby(['weather_main'])[['traffic_volume']].agg(np.sum)
Weather
```

#### Creating a Scatter Plot

```python
x = list(Weather.index) # With long holiday names, this column displays better on the y-axis.
y = list(Weather.traffic_volume) # This axis displays the traffic volume. 

fig, ax = plt.subplots(figsize=(15,8))
fig.subplots_adjust(bottom=0.2) # I have been running into images being cropped out when saved. 
# This above line provides padding to the figure so that all labels and axes are visible.  
ax.scatter(x, y)

ax.set_title('Traffic Volume in Different Weather Conditions', fontsize=18)
ax.set_xlabel('Traffic Volume', fontsize=12)
ax.set_ylabel('Weather', fontsize=12)
ax.tick_params(axis='x', labelrotation=90) 
ax.set_facecolor('None') 
ax.xaxis.grid(False)
ax.yaxis.grid(False)

fig.savefig('ThirdPlot.jpg', dpi=300)
```
